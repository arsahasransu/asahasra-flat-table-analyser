```python
# Estimate the number of TkEls lost in the DYPU200 sample that are present in MB and DYnoPU

bcDYPU = [0.0, 0.002, 0.007, 0.02, 0.038, 0.062, 0.085, 0.101, 0.109, 0.109, 0.1, 0.088, 0.28]
bcnoPU = [0.066, 0.376, 0.38, 0.118, 0.037, 0.013, 0.005, 0.002, 0.001, 0.0, 0.0, 0.0, 0.0]
bcMB = [0.003, 0.013, 0.032, 0.058, 0.085, 0.105, 0.116, 0.115, 0.107, 0.092, 0.075, 0.058, 0.141]
total_electron_count = 0
for i in range (13, 25):
    prob_tuple = [round(bcMB[i-k]*bcnoPU[k], 5) for k in range(1,13) if i-k<=12]
    prob = round(sum(prob_tuple), 3)
    total_electron_count += (i-12)*prob
    print("#", i, prob, prob_tuple, i-12, round(total_electron_count, 3))

print("#", round(total_electron_count, 3))
```
